## Create a Unique ID for Dataframe

```python
leave4[['str_Year', 'str_Month', 'str_Day']] = leave4.DAYOFSERVICE.str.split('-', expand=True)
```

```python
leave4.head()
```

```python
# Convert DAYOFSERVICE Column back to datetime for kernel space
leave4['DAYOFSERVICE'] = leave4['DAYOFSERVICE'].astype('datetime64')
```

```python
leave4['str_progrnumber'] = leave4['PROGRNUMBER'].astype('string')
```

```python
leave4.head()
```

```python
# Replace any under 10 stop with a leading zero
leave4['str_progrnumber'] = np.where(leave4['str_progrnumber'] == '1', '01', leave4['str_progrnumber'])
leave4['str_progrnumber'] = np.where(leave4['str_progrnumber'] == '2', '02', leave4['str_progrnumber'])
leave4['str_progrnumber'] = np.where(leave4['str_progrnumber'] == '3', '03', leave4['str_progrnumber'])
leave4['str_progrnumber'] = np.where(leave4['str_progrnumber'] == '4', '04', leave4['str_progrnumber'])
leave4['str_progrnumber'] = np.where(leave4['str_progrnumber'] == '5', '05', leave4['str_progrnumber'])
leave4['str_progrnumber'] = np.where(leave4['str_progrnumber'] == '6', '06', leave4['str_progrnumber'])
leave4['str_progrnumber'] = np.where(leave4['str_progrnumber'] == '7', '07', leave4['str_progrnumber'])
leave4['str_progrnumber'] = np.where(leave4['str_progrnumber'] == '8', '08', leave4['str_progrnumber'])
leave4['str_progrnumber'] = np.where(leave4['str_progrnumber'] == '9', '09', leave4['str_progrnumber'])
```

```python
leave4['id'] = leave4['str_Year'] + leave4['str_Month'] + leave4['str_Day'] + leave4['TRIPID'].map(str) + leave4['str_progrnumber']
```

```python
leave4.head()
```

```python
# Drop str_progrnumber Column for kernel space
leave4.drop('str_progrnumber', axis=1, inplace=True)
```

```python
leave4.head()
```

```python
leave4 = leave4.sort_values(by=['id'])
```

```python
leave4.head()
```
