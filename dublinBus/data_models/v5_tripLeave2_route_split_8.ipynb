```python
# Library Imports.
import pandas as pd

import numpy as np
```

```python
# Read tripleave2 route split 8 CSV file into Data Frame:
wk = pd.read_csv('v5_tripLeave2_route_split_8.csv', keep_default_na=True, delimiter=',', skipinitialspace=True)
```

```python
wk.shape
```

```python
wk.head()
```

```python
wk.dtypes
```

## Datatype Changes

```python
wk['num_lineID'] = wk['num_lineID'].astype('int16')
wk['DIRECTION'] = wk['DIRECTION'].astype('int16')
```

```python
wk['actual_duration'] = wk['actual_duration'].astype('int64')
wk['dayOfWeek'] = wk['dayOfWeek'].astype('int16')
```

```python
wk['weekend'] = wk['weekend'].astype('int16')
wk['rushHour'] = wk['rushHour'].astype('int16')
```

```python
wk.head()
```

```python
wk.dtypes
```

## Route 69

```python
lwk = wk[(wk['LINEID'] == '69')]
```

```python
lwk.shape
```

```python
# Drop Route 69 rows from wk
wk.drop(wk[wk['LINEID'] == '69'].index, inplace = True)
```

```python
wk.head()
```

```python
wk.shape
```

```python
# Send df to csv:
lwk.to_csv('v5_69_tripLeave2_sp8.csv', index=False)
```
