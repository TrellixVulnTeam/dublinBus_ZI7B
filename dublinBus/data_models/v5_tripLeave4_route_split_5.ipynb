```python
# Library Imports.
import pandas as pd

import numpy as np
```

```python
# Read tripleave4 route split 5 CSV file into Data Frame:
wk = pd.read_csv('v5_tripLeave4_route_split_5.csv', keep_default_na=True, delimiter=',', skipinitialspace=True)
```

```python
wk.shape
```

```python
wk.head()
```

```python
wk.dtypes
```

## Datatype Changes

```python
wk['num_lineID'] = wk['num_lineID'].astype('int16')
wk['DIRECTION'] = wk['DIRECTION'].astype('int16')
```

```python
wk['actual_duration'] = wk['actual_duration'].astype('int64')
wk['dayOfWeek'] = wk['dayOfWeek'].astype('int16')
```

```python
wk['weekend'] = wk['weekend'].astype('int16')
wk['rushHour'] = wk['rushHour'].astype('int16')
```

```python
wk.head()
```

```python
wk.dtypes
```

## Route 66A

```python
lwk = wk[(wk['LINEID'] == '66A')]
```

```python
lwk.shape
```

```python
# Drop Route 66A rows from wk
wk.drop(wk[wk['LINEID'] == '66A'].index, inplace = True)
```

```python
wk.head()
```

```python
wk.shape
```

<h3>Send to CSV</h3>

```python
# Send df to csv:
lwk.to_csv('v5_66A_tripLeave4_sp5.csv', index=False)
```
