## Read Route

```python
rt = pd.read_csv('v6_route_9.csv', keep_default_na=True, delimiter=',', skipinitialspace=True)
```

## Rename Columns for Clarity

```python
rt.rename(columns = {'actual_duration':'act_trip_dur'}, inplace = True)
```

```python
rt.head()
```

```python
rt.rename(columns = {'actualARR_hour':'stop_actARR_hour'}, inplace = True)
```

```python
rt.head()
```

## Create Weather Id in Route Data

```python
rt['DAYOFSERVICE'] = rt['DAYOFSERVICE'].astype('string')
```

```python
rt[['str_Year', 'str_Month', 'str_Day']] = rt.DAYOFSERVICE.str.split('-', expand=True)
```

```python
rt.head()
```

```python
rt['str_hour'] = rt['stop_actARR_hour'].astype('string')
```

```python
rt.head()
```

```python
rt['str_hour'].mask(rt['str_hour'] == '0', '00', inplace=True)
rt['str_hour'].mask(rt['str_hour'] == '1', '01', inplace=True)
rt['str_hour'].mask(rt['str_hour'] == '2', '02', inplace=True)
rt['str_hour'].mask(rt['str_hour'] == '3', '03', inplace=True)
rt['str_hour'].mask(rt['str_hour'] == '4', '04', inplace=True)
rt['str_hour'].mask(rt['str_hour'] == '5', '05', inplace=True)
rt['str_hour'].mask(rt['str_hour'] == '6', '06', inplace=True)
rt['str_hour'].mask(rt['str_hour'] == '7', '07', inplace=True)
rt['str_hour'].mask(rt['str_hour'] == '8', '08', inplace=True)
rt['str_hour'].mask(rt['str_hour'] == '9', '09', inplace=True)
```

```python
rt.head()
```

```python
rt['wthr_tr_lt_id'] = rt['str_Year'] + rt['str_Month'] + rt['str_Day'] + rt['str_hour']
```

```python
rt.head()
```

```python
rt.drop('str_Year', axis=1, inplace=True)
rt.drop('str_Month', axis=1, inplace=True)
rt.drop('str_Day', axis=1, inplace=True)
rt.drop('str_hour', axis=1, inplace=True)
```

```python
rt.head()
```

```python
rt.dtypes
```

```python
rt['wthr_tr_lt_id'] = rt['wthr_tr_lt_id'].astype('int64')
```

## Combine Route & Weather Data

```python
rtWeather = pd.merge(rt, weather, how="left", on=["wthr_tr_lt_id"])
```

```python
rtWeather.shape
```

## Send to CSV for Modeling

```python
# Send df to csv:
rtWeather.to_csv('v7_route_9.csv', index=False)
```
