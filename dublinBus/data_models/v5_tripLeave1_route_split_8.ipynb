```python
# Library Imports.
import pandas as pd

import numpy as np
```

```python
# Read tripleave1 route split 8 CSV file into Data Frame:
wk = pd.read_csv('v5_tripLeave1_route_split_8.csv', keep_default_na=True, delimiter=',', skipinitialspace=True)
```

```python
wk.head()
```

```python
wk.dtypes
```

## Datatype Changes

```python
wk['num_lineID'] = wk['num_lineID'].astype('int16')
wk['DIRECTION'] = wk['DIRECTION'].astype('int16')
```

```python
wk['actual_duration'] = wk['actual_duration'].astype('int64')
wk['dayOfWeek'] = wk['dayOfWeek'].astype('int16')
```

```python
wk['weekend'] = wk['weekend'].astype('int16')
wk['rushHour'] = wk['rushHour'].astype('int16')
```

```python
wk.head()
```

```python
wk.dtypes
```
