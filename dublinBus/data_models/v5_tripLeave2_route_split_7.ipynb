```python
# Library Imports.
import pandas as pd

import numpy as np
```

```python
# Read tripleave2 route split 7 CSV file into Data Frame:
wk = pd.read_csv('v5_tripLeave2_route_split_7.csv', keep_default_na=True, delimiter=',', skipinitialspace=True)
```

```python
wk.shape
```

```python
wk.head()
```

```python
wk.dtypes
```

## Datatype Changes

```python
wk['num_lineID'] = wk['num_lineID'].astype('int16')
wk['DIRECTION'] = wk['DIRECTION'].astype('int16')
```

```python
wk['actual_duration'] = wk['actual_duration'].astype('int64')
wk['dayOfWeek'] = wk['dayOfWeek'].astype('int16')
```

```python
wk['weekend'] = wk['weekend'].astype('int16')
wk['rushHour'] = wk['rushHour'].astype('int16')
```

```python
wk.head()
```

```python
wk.dtypes
```

## Route 39

```python
lwk = wk[(wk['LINEID'] == '39')]
```

```python
lwk.shape
```
