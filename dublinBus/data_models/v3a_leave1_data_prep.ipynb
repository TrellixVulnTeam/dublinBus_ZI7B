## Create a Unique ID for Dataframe

```python
leave1[['str_Year', 'str_Month', 'str_Day']] = leave1.DAYOFSERVICE.str.split('-', expand=True)
```

```python
leave1.head()
```

```python
# Convert DAYOFSERVICE Column back to datetime for kernel space
leave1['DAYOFSERVICE'] = leave1['DAYOFSERVICE'].astype('datetime64')
```

```python
leave1['str_progrnumber'] = leave1['PROGRNUMBER'].astype('string')
```

```python
leave1.head()
```

```python
# Replace any under 10 stop with a leading zero
leave1['str_progrnumber'] = np.where(leave1['str_progrnumber'] == '1', '01', leave1['str_progrnumber'])
leave1['str_progrnumber'] = np.where(leave1['str_progrnumber'] == '2', '02', leave1['str_progrnumber'])
leave1['str_progrnumber'] = np.where(leave1['str_progrnumber'] == '3', '03', leave1['str_progrnumber'])
leave1['str_progrnumber'] = np.where(leave1['str_progrnumber'] == '4', '04', leave1['str_progrnumber'])
leave1['str_progrnumber'] = np.where(leave1['str_progrnumber'] == '5', '05', leave1['str_progrnumber'])
leave1['str_progrnumber'] = np.where(leave1['str_progrnumber'] == '6', '06', leave1['str_progrnumber'])
leave1['str_progrnumber'] = np.where(leave1['str_progrnumber'] == '7', '07', leave1['str_progrnumber'])
leave1['str_progrnumber'] = np.where(leave1['str_progrnumber'] == '8', '08', leave1['str_progrnumber'])
leave1['str_progrnumber'] = np.where(leave1['str_progrnumber'] == '9', '09', leave1['str_progrnumber'])
```

```python
leave1['id'] = leave1['str_Year'] + leave1['str_Month'] + leave1['str_Day'] + leave1['TRIPID'].map(str) + leave1['str_progrnumber']
```

```python
leave1.head()
```

```python
# Drop str_progrnumber Column for kernel space
leave1.drop('str_progrnumber', axis=1, inplace=True)
```

```python
leave1.head()
```

```python
leave1 = leave1.sort_values(by=['id'])
```

```python
leave1.head()
```
